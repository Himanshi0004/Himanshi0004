```python
from sklearn.preprocessing import LabelEncoder
LE=LabelEncoder()

STYLE=LE.fit_transform(dresses_data['Style'])
dresses_data.drop('Style',axis=1,inplace=True)
dresses_data['Style']=STYLE


PRICE=LE.fit_transform(dresses_data['Price'])
dresses_data.drop('Price',axis=1,inplace=True)
dresses_data['Price']=PRICE

SIZE=LE.fit_transform(dresses_data['Size'])
dresses_data.drop('Size',axis=1,inplace=True)
dresses_data['Size']=SIZE


SEASON=LE.fit_transform(dresses_data['Season'])
dresses_data.drop('Season',axis=1,inplace=True)
dresses_data['Season']=SEASON

NECK_LINE=LE.fit_transform(dresses_data['NeckLine'])
dresses_data.drop('NeckLine',axis=1,inplace=True)
dresses_data['NeckLine']=NECK_LINE

SLEEVE_LENGTH=LE.fit_transform(dresses_data['SleeveLength'])
dresses_data.drop('SleeveLength',axis=1,inplace=True)
dresses_data['SleeveLength']=SLEEVE_LENGTH

WAISELINE=LE.fit_transform(dresses_data['waiseline'])
dresses_data.drop('waiseline',axis=1,inplace=True)
dresses_data['waiseline']=WAISELINE

MATERIAL=LE.fit_transform(dresses_data['Material'])
dresses_data.drop('Material',axis=1,inplace=True)
dresses_data['Material']=MATERIAL


FABRIC_TYPE=LE.fit_transform(dresses_data['FabricType'])
dresses_data.drop('FabricType',axis=1,inplace=True)
dresses_data['FabricType']=FABRIC_TYPE

DECORATION=LE.fit_transform(dresses_data['Decoration'])
dresses_data.drop('Decoration',axis=1,inplace=True)
dresses_data['Decoration']=DECORATION

PATTERN_TYPE=LE.fit_transform(dresses_data['Pattern Type'])
dresses_data.drop('Pattern Type',axis=1,inplace=True)
dresses_data['Pattern Type']=PATTERN_TYPE
```

```python
dresses_data.drop('index',axis=1,inplace=True)
```

```python
x=dresses_data[['Style','Price','Rating','Size','Season','NeckLine','SleeveLength','waiseline','Material','FabricType','Decoration','Pattern Type']]
y=dresses_data[['Recommendation']]

# Model Selection 

from sklearn.model_selection import train_test_split
from sklearn import metrics
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25)
print('xtrain:\n',xtrain)
print('xtest:\n',xtest)
print('ytrain:\n',ytrain)
print('ytest:\n',ytest)


#Feature selection with univariate stastical tests 
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif

Test= SelectKBest(score_func=f_classif, k=4)
fit = Test.fit(x,y)
print('fit score:\n',fit.scores_)
features = fit.transform(x)
print('features:\n',features[2:14,:])

#Feature scaling 
from sklearn.preprocessing import StandardScaler
S_scaling = StandardScaler()
xtrain_ss = S_scaling.fit_transform(xtrain)
xtest_ss = S_scaling.transform(xtest)
print(xtrain_ss)
print(xtest_ss)
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt 
%matplotlib inline

#correlation matrix

cor=dresses_data.corr()
print('correlation_matrix\n',cor)

#plotting Heatmap 
plt.figure(figsize=(10,6))
sns.heatmap(cor,annot=True)
```

```python
#using K-NN classifier
from sklearn.neighbors import KNeighborsClassifier

KNN_clf = KNeighborsClassifier(n_neighbors=11)
KNN_fit = KNN_clf.fit(xtrain, ytrain.values.ravel())

#predicting test result of KNN

KNN_y_pred=KNN_clf.predict(xtest)

#Feature selection
from sklearn.feature_selection import SequentialFeatureSelector
KNN_sfs=SequentialFeatureSelector(KNN_clf,n_features_to_select=2)
KNN_sfs.fit(x,y)

# confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(ytest, KNN_y_pred)
print ('confusion_matrix:',confusion_matrix)

#Model accuracy
KNN_Accuracy=metrics.accuracy_score(ytest,KNN_y_pred)

print('**KNN classifier**:\n',KNN_clf)
print('\n**KNN model fit**:\n  ',KNN_fit)
print('\n**KNN Test Prediction**:\n',KNN_y_pred)
print('\nFeatures selected by forward sequential selection:\n'f"{x.columns[KNN_sfs.get_support()].tolist()}")
print('\n**ACCURACY MODEL K-NN**:\n',KNN_Accuracy*100)

print('MODEL Score:\n',KNN_clf.score(x,y)*100)
```

```python
# using Random Forest Classifier

from sklearn.ensemble import RandomForestClassifier
Rf_clf=RandomForestClassifier(n_estimators=10,criterion='entropy')
Rf_clf.fit(xtrain,ytrain)

#predicting test model 

Rf_y_pred=Rf_clf.predict(xtest)



#features selection 
from sklearn.feature_selection import SequentialFeatureSelector

Rf_sfs=SequentialFeatureSelector(Rf_clf,n_features_to_select=2)
Rf_sfs.fit(x,y)

#finding model accuracy 

Rf_Accuracy=metrics.accuracy_score(ytest,Rf_y_pred)

print('\nRandom Forest Classifier:\n',Rf_clf)
print('\nRandom forest Test prediction:\n',Rf_y_pred)
print('\nFeatures selected by forward sequential selection:\n'f"{x.columns[Rf_sfs.get_support()].tolist()}")
print('\nACCURACY MODEL of Random forest:\n',Rf_Accuracy*100)

print('MODEL Score:\n',Rf_clf.score(x,y)*100)
```

```python
#using Decision Tree Classifier 

from sklearn.tree import DecisionTreeClassifier

#create Decision tree classifier object 
DT_clf=DecisionTreeClassifier()
DT_clf.fit(xtrain,ytrain)

#predict for test data 
DT_y_pred=DT_clf.predict(xtest)

#Confusion Matrix

from sklearn.metrics import confusion_matrix
confusion_matrix_DT = confusion_matrix(ytest, DT_y_pred)
print ('confusion_matrix:',confusion_matrix_DT)

#feature selection 

DT_sfs=SequentialFeatureSelector(DT_clf,n_features_to_select=2)
DT_sfs.fit(x,y)

print('Decision Tree Classifier:\n',DT_clf)
print('\nDecision Tree Test prediction:\n',DT_y_pred)
print('\nFeatures selected by forward sequential selection:\n'f"{x.columns[DT_sfs.get_support()].tolist()}")

#model accuracy
print('\nMODEL ACCURACY Decision Tree:\n',metrics.accuracy_score(ytest,DT_y_pred)*100)

print('Model Score:\n',DT_clf.score(x,y)*100)
```
